```python
# Visualizziamo lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: quali sono le differenze? C'è un pattern? C'è qualche elemento che non lo segue?
df.tail(20)[["StandardCost", "DealerPrice"]]
```

Il dataset diabetes.csv raccoglie persone con diabete o meno, e il valore di diverse variabili fisiologiche dei pazienti. 
• Osserviamone le dimensioni e un'anteprima di cinque righe; 
• Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne; 
• Stampiamo dei descrittori statistici del dataset; 
• Selezioniamo i dati relativi a diverse fasce di età: 20-29, 30-39, 40-50; 
• Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?

```python
import pandas as pd

df_dia = pd.read_csv("diabetes.csv")
df_dia.head()
```

```python
# Osserviamone le dimensioni e un'anteprima di cinque righe
df_dia.shape
df_dia.head()
```

```python
# Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne
df_dia.info()
```

```python
# Stampiamo dei descrittori statistici del dataset
df_dia.describe()
```

```python
# Selezioniamo i dati relativi a diverse fasce di età: 20-29, 30-39, 40-50
fascia_20_29 = df_dia[(df_dia["Age (years)"] >= 20) & (df_dia["Age (years)"] <= 29)]
fascia_30_39 = df_dia[(df_dia["Age (years)"] >= 30) & (df_dia["Age (years)"] <= 39)]
fascia_40_50 = df_dia[(df_dia["Age (years)"] >= 40) & (df_dia["Age (years)"] <= 50)]
```

```python
# Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?
fascia_20_29["Diastolic blood pressure (mm Hg)"].mean()
fascia_30_39["Diastolic blood pressure (mm Hg)"].mean()
fascia_40_50["Diastolic blood pressure (mm Hg)"].mean()
```

Il dataset insurance.csv contiene dati rispetto a caratteristiche e abitudini delle persone, 
e della zona in cui vivono, rispetto ai costi individuali per le cure mediche come premio per le assicurazioni sulla salute. 
• Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
• Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 
• E rispetto a smoker? E a sex? 
• Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?

```python
import pandas as pd

df_ins = pd.read_csv("insurance.csv")
```

```python
# Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna
df_ins.shape
df_ins.head()
df_ins.columns
```

```python
# Quali sono le medie di charges rispetto a region? Ci sono differenze significative?
df_ins.groupby("region")["charges"].mean()
```

```python
# E rispetto a smoker? 
df_ins.groupby("smoker")["charges"].mean()
```

```python
# E a sex?
df_ins.groupby("sex")["charges"].mean()
```

```python
# Quali sono i descrittori statistici di bmi?
df_ins["bmi"].describe()
```

```python
# Quali sono i descrittori statistici di bmi? 
df_ins["bmi_quartile"] = pd.qcut(df_ins["bmi"], 4)
```

```python
# Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
df_ins.groupby("bmi_quartile")["charges"].agg(["min", "mean", "max"])
```

Il dataset pokemon.csv contiene un database di Pokémon, con dati quali nome, tipi di attacco, valori di attacco/difesa/eccetera, e se sono o meno leggendari. 
• Verifichiamo la dimensione, un'anteprima e osserviamo i nomi di colonna; 
• È verosimile che la prima colonna dovrebbe essere un indice? 
• Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano? 
• Se no, settare la prima colonna come indice.

```python
# Verifichiamo la dimensione, un'anteprima e osserviamo i nomi di colonna
import pandas as pd

df_poke = pd.read_csv("pokemon.csv")

df_poke.shape
df_poke.head()
df_poke.columns
```

```python
# È verosimile che la prima colonna dovrebbe essere un indice?
df_poke.iloc[:10, 0]
```

```python
# Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano?
df_poke.index[:10]
```

```python
# Se no, settare la prima colonna come indice.
df_poke = pd.read_csv("pokemon.csv", index_col=0)
df_poke.head()
```

```python
# Quali sono i Pokémon leggendari?
df_poke[df_poke["Legendary"] == True]
```

```python
# E quali sono i leggendari di tipo 1 Grass?
df_poke[(df_poke["Legendary"] == True) & (df_poke["Type 1"] == "Grass")]
```

```python
# E leggendari di tipo 1 Ice o Fire?
df_poke[(df_poke["Legendary"] == True) & (df_poke["Type 1"].isin(["Ice", "Fire"]))]
```

```python
# Trasformiamo la colonna Name nell'indice
df_poke = df_poke.set_index("Name")
df_poke.head()
```

```python
# Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
df_poke[
    (df_poke["Generation"] == 1) &
    (df_poke["Attack"] > 50) &
    (df_poke["HP"] < 60)
]
```

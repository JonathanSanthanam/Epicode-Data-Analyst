## Calcola la somma cumulativa dei guadagni utilizzando il metodo .cumsum()

```python
import pandas as pd
import numpy as np

years = 5
guadagni = pd.DataFrame({
    "Mese": list("GFMAMGLASOND") * years,   # 12 mesi ripetuti
    "Anno": np.repeat(list(range(1, years+1)), 12),  # 1..5
    "Valore_Guadagni_Mensili": np.random.randint(800, 5000, 12*years)
})

guadagni["SumCumulativa"] = guadagni["Valore_Guadagni_Mensili"].cumsum()
print(guadagni)
```

## Come sopra, ma raggruppato per ogni anno usando prima un .groupby()

```python
import pandas as pd
import numpy as np

years = 5
guadagni = pd.DataFrame({
    "Mese": list("GFMAMGLASOND") * years,   # 12 mesi ripetuti
    "Anno": np.repeat(list(range(1, years+1)), 12),  # 1..5
    "Valore_Guadagni_Mensili": np.random.randint(800, 5000, 12*years)
})

guadagni["SumCumulativa"] = guadagni.groupby("Anno")["Valore_Guadagni_Mensili"].cumsum()
print(guadagni)
```

## Dai beginner_datasets carichiamo in un DataFrame il file facebook.csv, che contiene dei post con data di pubblicazione, tipo (foto, video, …) e numero di reactions raccolte:

```python
import pandas as pd

df_fb = pd.read_csv("../beginner_datasets/facebook.csv")
print(df_fb.head())
```

## Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp

```python
df_fb["Timestamp"] = pd.to_datetime(df_fb["status_published"])
print(df_fb.head())
```

## Utilizzando gli attributi .dt.year , .dt.month , .dt.day , .dt.dayofweek , .dt.dayofyear, ottieniamo informazioni specifiche sulle date delle transazioni, come l'anno, il mese, il giorno della settimana, il giorno dell'anno, eccetera

```python
df_fb["Timestamp"] = pd.to_datetime(df_fb["status_published"])
df_fb["Year"] = df_fb["Timestamp"].dt.year
df_fb["Month"] = df_fb["Timestamp"].dt.month
df_fb["Day"] = df_fb["Timestamp"].dt.day
df_fb["Day_of_Week"] = df_fb["Timestamp"].dt.dayofweek
print(df_fb[["Year", "Month", "Day", "Day_of_Week"]].head())
```

## Estraiamo solo i post relativi al 2012

```python
df_2012 = df_fb[df_fb["Year"] == 2012]
print(df_2012.head())
```

## Estraiamo solo i post relativi a maggio 2018

```python
df_maggio_2018 = df_fb[(df_fb["Year"] == 2018) & (df_fb["Month"] == 5)]
print(df_maggio_2018.head())
```

## Confrontiamo il numero di post pubblicati nei weekend rispetto al numero di post pubblicati nel resto della settimana

```python
weekend_posts = df_fb[df_fb["Day_of_Week"].isin([5, 6])]
weekday_posts = df_fb[~df_fb["Day_of_Week"].isin([5, 6])]

print("Posts weekend:", len(weekend_posts))
print("Posts weekdays:", len(weekday_posts))
```

## Troviamo il primo e ultimo post pubblicati in ogni anno

```python
df_fb["Timestamp"] = pd.to_datetime(df_fb["status_published"])
df_fb["Year"] = df_fb["Timestamp"].dt.year

primi_ultimi = (df_fb.groupby("Year")["Timestamp"].agg(["min", "max"]))

print(primi_ultimi)
```

Dovremo contare i dati nulli, utilizzando .sum(); potremmo doverlo utilizzare più di una volta

```python
import os
import pandas as pd

path = "../beginner_datasets/"

for file in os.listdir(path):
    if file.endswith(".csv"):
        df = pd.read_csv(path + file)
        mask_null = df.isna()

        null_per_col = mask_null.sum()
        total_null = null_per_col.sum()
```

Dovremo stampare, o memorizzare in una list, solo i nomi dei file che contengono dati nulli

```python
import os
import pandas as pd

path = "../beginner_datasets/"
dataset_con_nulli = []

for file in os.listdir(path):
    if file.endswith(".csv"):
        df = pd.read_csv(path + file)
        mask_null = df.isna()
        total_null = mask_null.sum().sum()

        if total_null > 0:
            dataset_con_nulli.append(file)

print(dataset_con_nulli)
```
